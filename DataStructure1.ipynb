## Linked List

Unlike an array linked list will not have a fixed size.

Each node in a linked list has a reference to the next node in the list.
End nodes do not have a reference to the next node.
Linked list can be traversed from the beginning to the end or from the end to the beginning depending on the traversal direction.
Such as:

```python
# Forward traversal
node = head
while node:
    print(node.data)
    node = node.next

# Backward traversal
node = tail
while node:
    print(node.data)
    node = node.prev
```

To do this we need to keep track of the previous node.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
```




````python
class Element(object):
    def __init__(self, value):
        self.value = value
        self.next = None
        
class LinkedList(object):
    def __init__(self, head=None):
        self.head = head
        
    def append(self, new_element):
        current = self.head
        if self.head:
            while current.next:
                current = current.next
            current.next = new_element
        else:
            self.head = new_element
            
    def get_position(self, position):
        """Get an element from a particular position.
        Assume the first position is "1".
        Return "None" if position is not in the list.
        Return the element if found."""
        current = self.head
        count = 1
        while current:
            if count == position:
                return current
            current = current.next
            count += 1
        return None

    
    def insert(self, new_element, position):
        """Insert a new node at the given position.
        Assume the first position is "1".
        Inserting at position 3 means between
        the 2nd and 3rd elements."""
        current = self.head
        count = 1
        while current:
            if count == position - 1:
                new_element.next = current.next
                current.next = new_element
                return
            current = current.next
            count += 1
        
    
    def delete(self, value):
        """Delete the first node in the linked list with a given value."""
        current = self.head
        if current.value == value:
            self.head = current.next
            return
        while current:
            if current.next.value == value:
                current.next = current.next.next
                return
            current = current.next

# Test cases
# Set up some Elements
e1 = Element(1)
e2 = Element(2)
e3 = Element(3)
e4 = Element(4)

# Start setting up a LinkedList
ll = LinkedList(e1)
ll.append(e2)
ll.append(e3)

# Test get_position
# Should print 3
print(ll.head.next.next.value)
# Should also print 3
print(ll.get_position(3).value)

# Test insert
ll.insert(e4,3)
# Should print 4 now
print(ll.get_position(3).value)

# Test delete
ll.delete(1)
# Should print 2 now
print(ll.get_position(1).value)
# Should print 4 now
print(ll.get_position(2).value)
# Should print 3 now
print(ll.get_position(3).value)
````

## Queues

Queues are a data structure that operate in a FIFO (first in first out) fashion.
Meaning that the first item that is added to the queue is the first item that is removed.

```python
class Queue:
    def __init__(self):
        self.items = []
    
    def enqueue(self, item):
        self.items.insert(0, item)
    
    def dequeue(self):
        return self.items.pop()
    
    def is_empty(self):
        return self.items == []
    
    def size(self):
        return len(self.items)
```

````python
class Queue:
    def __init__(self, head=None):
        self.storage = [head]
    
    def enqueue(self, new_element):
        self.storage.append(new_element)
    
    def peek(self):
        return self.storage[0]
    
    def dequeue(self):
        return self.storage.pop(0)
````

## Binary Search

Is a search algorithm that works by comparing the target value to the middle element of the list.
If the target value is less than the middle element, then the algorithm repeats the search on the lower half of the list.
If the target value is greater than the middle element, then the algorithm repeats the search on the upper half of the list.
If the target value is equal to the middle element, then the target value is found and the algorithm terminates.

**The binary search needs a sorted list.**

This is a O(log n) algorithm.

### Results Table for Binary Search
| Array Size | Iterations | Power of 2 |
| -----------| -----------| -----------|
| 0          | 0          | $2^0$          |
| 1          | 1          | $2^0$          |
| 2          | 2          | $2^1$          |
| 3          | 2          | $2^1$          |
| 4          | 3          | $2^2$          |
| 5          | 3          | $2^2$          |
| 6          | 3          | $2^2$          |
| 7          | 3          | $2^2$          |
| 8          | 4          | $2^3$          |


This leads to the $O(\log(n))$ algorithm.

````python
def binarySearch(input_array, value):
    left = 0
    right = len(input_array) - 1
    while left <= right:
        mid = (left + right) // 2
        if input_array[mid] == value:
            return mid
        elif input_array[mid] < value:
            left = mid + 1
        else:
            right = mid - 1
    return -1

test_array = [1,3,9,11,15,17,19,21,25]

print(binarySearch(test_array, 20))
````

## Recursion

Recursion is a technique where a function calls itself.
Will have a base case and a recursive case.

```python
def factorial(n):
    if n == 0:
        return 1 # base case
    else:
        return n * factorial(n-1) # recursive case
```

````python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(7))
````

## Sorting

Naive sorting algorithm is the simplest sorting algorithm.
It compares each element with every other element and swaps them if they are in the wrong order.

Sorting can be in place or not in place.
In place sorting is when the list is sorted in place.
Not in place sorting is when the list is sorted and returned as a new list.

In-place sorting is the most efficient sorting algorithm, but not the most space efficient.

### Bubble Sort

This is a Naive sorting algorithm.

It compares each element with every other element and swaps them if they are in the wrong order.  
The Largest element will bubble to the end of the list.  
This algorithm has an worst case of O(n^2).  


```python
def bubble_sort(items):
    for i in range(len(items)-1):
        for j in range(len(items)-1-i):
            if items[j] > items[j+1]:
                items[j], items[j+1] = items[j+1], items[j]
    return items
```

### Merge Sort

This is a Divide and Conquer sorting algorithm.
It recursively splits the list into smaller sublists until the sublists are of size 1.
Then the sublists are sorted and merged back together.
This algorithm has an average case of O(n log n) and a worst case of O(n log n).

```python
def merge_sort(items):
    if len(items) > 1:
        mid = len(items) // 2
        left = items[:mid]
        right = items[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                items[k] = left[i]
                i += 1
            else:
                items[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            items[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            items[k] = right[j]
            j += 1
            k += 1
    return items
```

### Quick Sort

We will be using the last element of the list as the pivot.
If the pivot is the largest element, then the algorithm will recursively sort the list from the first element to the pivot.
If the pivot is the smallest element, then the algorithm will recursively sort the list from the pivot to the last element.
This algorithm has an average case of O(n log n) and a worst case of O(n^2).

```python
def quick_sort(items):
    if len(items) > 1:
        pivot = items[-1]
        i = 0
        for j in range(len(items)-1):
            if items[j] < pivot:
                items[i], items[j] = items[j], items[i]
                i += 1
        items[i], items[-1] = items[-1], items[i]
        quick_sort(items[:i])
        quick_sort(items[i+1:])
    return items
```

````python
def quickSort(array):
    if len(array) <= 1:
        return array
    else:
        pivot = array[0]
        less = [i for i in array[1:] if i <= pivot]
        greater = [i for i in array[1:] if i > pivot]
        return quickSort(less) + [pivot] + quickSort(greater)

unsortedList = [21,4,1,3,9,20,25]

print(quickSort(unsortedList))
````
